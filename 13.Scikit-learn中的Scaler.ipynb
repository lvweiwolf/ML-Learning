# Scikit-learn中的Scaler

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
Y = iris.target
X[:10, :]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=666)
```

## scikit-learn中的StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdScaler = StandardScaler()
stdScaler.fit(X_train)
```

```python
stdScaler.mean_
```

```python
stdScaler.scale_
```

```python
X_train_standard = stdScaler.transform(X_train)
X_train_standard
```

```python
X_test_standard = stdScaler.transform(X_test)
X_test_standard
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
knn_clf.fit(X_train_standard, Y_train)
```

```python
knn_clf.score(X_test_standard, Y_test)
```

```python
knn_clf.score(X_test, Y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(X_train)
```

```python
min_max_scaler.min_
```

```python
min_max_scaler.data_max_
```

```python
X_train_min_max = min_max_scaler.transform(X_train)
X_train_min_max
```

```python
X_test_min_max = min_max_scaler.transform(X_test)
X_test_min_max
```

```python
knn_clf2 = KNeighborsClassifier(n_neighbors=3)
knn_clf2.fit(X_train_min_max, Y_train)
```

```python
knn_clf2.score(X_test_min_max, Y_test)
```
