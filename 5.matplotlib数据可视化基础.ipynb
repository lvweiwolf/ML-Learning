# matplotlib 基础

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
help(plt.plot)
```

```python
siny = y.copy()
cosy = np.cos(x)
plt.plot(x, siny)
plt.plot(x, cosy, color='r', linestyle='--')
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='r', linestyle='--')
plt.xlim(-5.0, 15.0)
plt.ylim(0.0, 1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='r', linestyle='--')
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='r', linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color='r', linestyle='--', label="cos(x)")
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color='r', linestyle='--', label="cos(x)")
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.title('Welcome to the ML World!')
plt.show()
```

## Scatter Plot (散点图)

```python
plt.scatter(x, siny)
plt.show()
```
