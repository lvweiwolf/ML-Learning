# 梯度下降算法模拟

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 140)
plot_x
```

```python
plot_y = (plot_x - 2.5)**2 - 1
```

```python
plt.plot(plot_x, plot_y)
```

```python
def dJ(theta):
    return 2 * (theta - 2.5)
```

```python
def J(theta):
    return (theta - 2.5)**2 - 1
```

```python
eta = 0.1
theta = 0.0
epsilon = 1e-8

while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta*gradient

    if(abs(J(theta) - J(last_theta)) < epsilon):
        break
    
print(theta)
print(J(theta))
```

```python
theta = 0.0
theta_history = [theta]

while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta*gradient
    theta_history.append(theta)

    if(abs(J(theta) - J(last_theta)) < epsilon):
        break
    
plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
plt.show()
```

```python
def gradient_descent(init_theta, eta, epsilon=1e-8):
    theta = init_theta
    theta_history.append(init_theta)

    while True:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta*gradient
        theta_history.append(theta)

        if(abs(J(theta) - J(last_theta)) < epsilon):
            break

def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
    plt.show()
```

```python
eta = 0.01
theta_history = []
gradient_descent(0., eta)
plot_theta_history()
```

```python
def J(theta):
    try: 
        return (theta - 2.5)**2 - 1.
    except:
        return float('inf')
```

```python
def gradient_descent(init_theta, eta, n_iters = 1e4, epsilon=1e-8):
    theta = init_theta
    theta_history.append(init_theta)
    i_iter = 0

    while i_iter < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta*gradient
        theta_history.append(theta)

        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
        
        i_iter += 1
```

```python
eta = 1.1
theta_history = []
gradient_descent(0., eta)
```
